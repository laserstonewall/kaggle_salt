```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['lines.linewidth'] = 2
matplotlib.rcParams['xtick.labelsize'] = 16
matplotlib.rcParams['ytick.labelsize'] = 16
matplotlib.rcParams['axes.labelsize'] = 20
matplotlib.rcParams['axes.titlesize'] = 20

%matplotlib inline
```

```python
df = pd.read_csv('./automated_parameter_search_1.csv')

# df.head()

agged = df.groupby(['filter_scaling','depth','batchsize']).agg(np.mean).reset_index()

agged.head(10)
```

```python
curr = agged[agged['filter_scaling']==8]; plt.plot(curr['depth'], curr['mAP_max'], label='8')
curr = agged[agged['filter_scaling']==16]; plt.plot(curr['depth'], curr['mAP_max'], label='16')
curr = agged[agged['filter_scaling']==32]; plt.plot(curr['depth'], curr['mAP_max'], label='32')
plt.legend(); plt.show()
curr = agged[agged['filter_scaling']==8]; plt.plot(curr['depth'], curr['mAP_max_augmentation'], label='8 aug')
curr = agged[agged['filter_scaling']==16]; plt.plot(curr['depth'], curr['mAP_max_augmentation'], label='16 aug')
curr = agged[agged['filter_scaling']==32]; plt.plot(curr['depth'], curr['mAP_max_augmentation'], label='32 aug')
plt.legend(); plt.show()
```

```python
curr = agged[agged['depth']==4]; plt.plot(curr['filter_scaling'], curr['mAP_max'], label='4')
curr = agged[agged['depth']==5]; plt.plot(curr['filter_scaling'], curr['mAP_max'], label='5')
curr = agged[agged['depth']==6]; plt.plot(curr['filter_scaling'], curr['mAP_max'], label='6')
plt.legend(); plt.show()
curr = agged[agged['depth']==4]; plt.plot(curr['filter_scaling'], curr['mAP_max_augmentation'], label='4 aug')
curr = agged[agged['depth']==5]; plt.plot(curr['filter_scaling'], curr['mAP_max_augmentation'], label='5 aug')
curr = agged[agged['depth']==6]; plt.plot(curr['filter_scaling'], curr['mAP_max_augmentation'], label='6 aug')
plt.legend(); plt.show()
```
