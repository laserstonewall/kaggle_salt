```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['lines.linewidth'] = 2
matplotlib.rcParams['xtick.labelsize'] = 16
matplotlib.rcParams['ytick.labelsize'] = 16
matplotlib.rcParams['axes.labelsize'] = 20
matplotlib.rcParams['axes.titlesize'] = 20

%matplotlib inline
```

```python
df = pd.read_csv('./automated_parameter_search_1.csv')

# df.head()

agged = df.groupby(['filter_scaling','depth','batchsize']).agg(np.mean).reset_index()

agged.head(20)
```

```python
curr = agged[agged['filter_scaling']==8]; plt.plot(curr['depth'], curr['mAP_max'], 'o-', label='8')
curr = agged[agged['filter_scaling']==16]; plt.plot(curr['depth'], curr['mAP_max'], 'o-', label='16')
curr = agged[agged['filter_scaling']==32]; plt.plot(curr['depth'], curr['mAP_max'], 'o-', label='32')
curr = agged[agged['filter_scaling']==48]; plt.plot(curr['depth'], curr['mAP_max'], 'o-', label='48')
curr = agged[agged['filter_scaling']==64]; plt.plot(curr['depth'], curr['mAP_max'], 'o-', label='64')
plt.legend(loc='lower right'); plt.show()
curr = agged[agged['filter_scaling']==8]; plt.plot(curr['depth'], curr['mAP_max_augmentation'], 'o-', label='8 aug')
curr = agged[agged['filter_scaling']==16]; plt.plot(curr['depth'], curr['mAP_max_augmentation'], 'o-', label='16 aug')
curr = agged[agged['filter_scaling']==32]; plt.plot(curr['depth'], curr['mAP_max_augmentation'], 'o-', label='32 aug')
curr = agged[agged['filter_scaling']==48]; plt.plot(curr['depth'], curr['mAP_max_augmentation'], 'o-', label='48 aug')
curr = agged[agged['filter_scaling']==64]; plt.plot(curr['depth'], curr['mAP_max_augmentation'], 'o-', label='64 aug')
plt.legend(loc='lower right'); plt.show()
```

```python
curr = agged[agged['depth']==3]; plt.plot(curr['filter_scaling'], curr['mAP_max'], 'o-', label='3')
curr = agged[agged['depth']==4]; plt.plot(curr['filter_scaling'], curr['mAP_max'], 'o-', label='4')
curr = agged[agged['depth']==5]; plt.plot(curr['filter_scaling'], curr['mAP_max'], 'o-', label='5')
curr = agged[agged['depth']==6]; plt.plot(curr['filter_scaling'], curr['mAP_max'], 'o-', label='6')
plt.legend(); plt.show()
curr = agged[agged['depth']==3]; plt.plot(curr['filter_scaling'], curr['mAP_max_augmentation'], 'o-', label='3 aug')
curr = agged[agged['depth']==4]; plt.plot(curr['filter_scaling'], curr['mAP_max_augmentation'], 'o-', label='4 aug')
curr = agged[agged['depth']==5]; plt.plot(curr['filter_scaling'], curr['mAP_max_augmentation'], 'o-', label='5 aug')
curr = agged[agged['depth']==6]; plt.plot(curr['filter_scaling'], curr['mAP_max_augmentation'], 'o-', label='6 aug')
plt.legend(); plt.show()
```

```python
curr = agged[agged['filter_scaling']==48]; plt.plot(curr['batchsize'], curr['mAP_max'], 'o-', label='48')
plt.legend(loc='lower right'); plt.show()
curr = agged[agged['filter_scaling']==48]; plt.plot(curr['batchsize'], curr['mAP_max_augmentation'], 'o-', label='48')
plt.legend(loc='lower right'); plt.show()
```
